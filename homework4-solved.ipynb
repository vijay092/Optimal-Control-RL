# Code Question

Consider the Linear dynamic system 

$$\dot x = A_{\mathrm{dyn}} x$$

where $A_{\mathrm{dyn}}$ and the initial condition are given below.

```python
A_dyn = np.array([[0,2,0],
                  [-2.,0.,0],
                  [10,0,-10]])

x0 = np.array([1.,0,0])
```

In this question, we will be computing the solution for $30$ equally spaced points over $[0,2\pi]$ using various collocation methods.

The exact solution for this system is given below.

```python
import scipy.linalg as la
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline



Time = np.linspace(0,2*np.pi,30)
x0 = np.array([1.,0.,0.])
X_exact = []


for t in Time:
    x = la.expm(A_dyn*t) @ x0
    X_exact.append(x)
    
X_exact = np.array(X_exact).T

n = len(x0)

fig,ax = plt.subplots(n,1,sharex=True)
ax[-1].set_xlabel('Time')
for i in range(n):
    ax[i].plot(Time,X_exact[i])
    ax[i].set_ylabel(r'$x_%d(t)$' % (i+1))
    
```

Use CasADi to simulate this system by collocation with step size $h=2\pi / 30$, for the following collocation vectors:

For each simulation plot it with the exact solution to get a feel for its accuracy.

```python
# Solution Code

import numpy as np
import scipy.linalg
import runge_kutta as rk
from casadi import *

C_vecs = [np.array([0.]),
          np.array([.5]),
          np.array([1.]),
          np.array([0.,1]),
          np.array(collocation_points(2,'legendre')),
          np.array([0,.5,1])]


f = lambda t,x : A_dyn @ x

for c in C_vecs:
    c,A,b = rk.collocation_tableau(c)

    X_RK = rk.runge_kutta_sim(f,Time,x0,c,A,b)
    
    fig,ax = plt.subplots(n,1,sharex=True)
    c_str = ','.join([str(ci) for ci in c])
    ax[0].set_title('c = [%s]' % c_str)
    ax[-1].set_xlabel('Time')
    for i in range(n):
        ax[i].plot(Time,X_exact[i])
        ax[i].plot(Time,X_RK[i],'o')
        ax[i].set_ylabel(r'$x_%d(t)$' % (i+1))
        
```
